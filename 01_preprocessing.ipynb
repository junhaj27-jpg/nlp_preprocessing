```python
import re
from konlpy.tag import Okt

okt = Okt()

```

```python
#  통합 불용어
stopwords = {
    "은","는","이","가","을","를","에","의","과","와","도",
    "하다","되다","이다","있다","수","더","통해",
    "그리고","하지만","또한","그것","이것","저것"
}

```

```python
#  반복 문자 정규화
def normalize_repeat(text):
    text = re.sub(r'(ㅋ)\1{2,}', r'\1\1', text)
    text = re.sub(r'(ㅎ)\1{2,}', r'\1\1', text)
    text = re.sub(r'(ㅠ)\1{2,}', r'\1\1', text)
    return text

```

```python
#  최종 전처리 함수
def preprocess(text):
    
    # 1️ 반복 문자 정규화
    text = normalize_repeat(text)
    
    # 2️ 소문자 변환
    text = text.lower()
    
    # 3️ 숫자 제거
    text = re.sub(r'\d+', ' ', text)
    
    # 4️ 특수문자 제거 (한글 + 영어만 유지)
    text = re.sub(r'[^가-힣a-z\s]', ' ', text)
    
    # 5️ 공백 정리
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 6️ 형태소 분석 + 표제어 추출
    tokens = okt.pos(text, stem=True)
    
    # 7️ 품사 필터링 + 불용어 제거
    tokens = [
        word for word, pos in tokens
        if pos in ["Noun", "Verb", "Adjective"]
        and word not in stopwords
        and len(word) > 1
    ]
    
    return tokens
```

```python
import re
from konlpy.tag import Okt

okt = Okt()

def preprocess_korean(text):
    text = re.sub(r"[^가-힣\s]", "", text)
    tokens = okt.pos(text, stem=True)
    tokens = [
        word for word, pos in tokens
        if pos in ["Noun", "Verb", "Adjective"]
        and len(word) > 1
    ]
    return tokens
```

```python
tokens_all = [preprocess_korean(sent) for sent in corpus_clean[:5000]]
print(tokens_all[:2])
```
