```python
import pickle

with open("../data/preprocessed_tokens.pkl", "rb") as f:
    tokens_all = pickle.load(f)

print(len(tokens_all))
print(tokens_all[:2])
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=tokens_all,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)

print(model.wv['한국'])
```

```python
import numpy as np

def sentence_vector(tokens):
    
    vectors = []
    
    for word in tokens:
        if word in model.wv:
            vectors.append(model.wv[word])
    
    if len(vectors) == 0:
        return np.zeros(100)
    
    return np.mean(vectors, axis=0)
```

```python
positive_words = ['좋다','재밌다','훌륭하다','최고']
negative_words = ['별로','지루하다','최악','실망']

pos_vec = sentence_vector(positive_words)
neg_vec = sentence_vector(negative_words)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def predict_sentiment(tokens):
    
    vec = sentence_vector(tokens)
    
    pos_sim = cosine_similarity([vec],[pos_vec])[0][0]
    neg_sim = cosine_similarity([vec],[neg_vec])[0][0]
    
    if pos_sim > neg_sim:
        return "긍정"
    else:
        return "부정"
```

```python
test_sentence = ['이','영화','정말','재밌다']

result = predict_sentiment(test_sentence)

print(result)
```
