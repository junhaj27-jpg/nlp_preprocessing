# Word Embedding을 활용한 단어 찾기 게임

- Level 1. [Semantle](https://semantle.com/) 흉내내기
- Level 2. Semantle의 한국어 버전을 만들어보기
- Level 3. 단어 `A:B=C:?` 여기서 ? 맞추기 ([한국어 임베딩 확인](https://word2vec.kr/search/) + Semantle)

```python
import numpy as np
from gensim.models import Word2Vec

# 예제 학습 데이터
sentences = [
    ["king","queen","man","woman"],
    ["paris","france","london","england"],
    ["apple","fruit","banana","orange"],
    ["dog","cat","animal"],
    ["car","bus","train","transport"]
]

# Word2Vec 모델 학습
model = Word2Vec(sentences, vector_size=50, window=2, min_count=1)

# 코사인 유사도
def cosine_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

target = "king"

while True:

    guess = input("추측 단어: ").strip()

    if guess not in model.wv or target not in model.wv:
        print("사전에 없는 단어")
        continue

    sim = cosine_sim(model.wv[guess], model.wv[target]) * 100

    print(f"유사도: {sim:.2f}")

    if guess == target:
        print("정답!")
        break
```

```python
import pickle

with open("../data/preprocessed_tokens.pkl", "rb") as f:
    tokens_all = pickle.load(f)

print(len(tokens_all))
print(tokens_all[:3])
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=tokens_all,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4,
    sg=1
)

print("단어 수:", len(model.wv))
```

```python
import numpy as np

def cosine_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
import random

target = random.choice(model.wv.index_to_key)

print("한국어 단어 찾기 게임 시작!")
print("(게임 종료: exit)")
# print("정답:", target)  # 테스트용
```

```python
history = []

while True:
    
    guess = input("추측 단어: ").strip()

    if guess == "exit":
        print("게임 종료")
        print("정답:", target)
        break

    if guess not in model.wv:
        print("사전에 없는 단어")
        continue

    sim = cosine_sim(model.wv[guess], model.wv[target]) * 100

    history.append((guess, sim))

    if sim > 80:
        hint = "🔥 매우 가까움"
    elif sim > 60:
        hint = "🌡️ 가까움"
    elif sim > 40:
        hint = "🙂 관련 있음"
    else:
        hint = "🧊 멀다"

    print(f"{guess} → 유사도 {sim:.2f} ({hint})")

    if guess == target:
        print("🎉 정답!")
        break
```

```python
A = "남자"
B = "왕"
C = "여자"

print(f"{A}:{B} = {C}: ?")

result = model.wv[B] - model.wv[A] + model.wv[C]

answer = model.wv.similar_by_vector(result, topn=5)

print("후보:", answer)
```
